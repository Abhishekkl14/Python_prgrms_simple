```python
 #Time series analysis with Pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
date_rng = pd.date_range(start = '2023-01-01', end='2023-12-31', freq='D')

np.random.seed(42)
data = np.random.rand(len(date_rng))*10+50
df = pd.DataFrame(data, columns=['values'], index=date_rng)

print("Sample Time Series data:\n", df.head())

plt.figure(figsize=(10,6))
plt.plot(df.index, df['values'], label='Daily Data')
plt.title("Daily time series data")
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_monthly = df.resample('M').mean()

print("\n Monthly Resampled Data:\n", df_monthly)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_monthly.index, df_monthly['values'], label='Monthly Average', color='orange')
plt.title('Monthly Average Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['rolling_mean'] = df['values'].rolling(window=30).mean()
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['values'], label='Daily Data')
plt.plot(df.index, df['rolling_mean'], label='30-Day Rolling Mean', color='red')
plt.title('Time Series with 30-Day Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
